```python
# Clase ciudad atributos; nombre, pais, poblacion
# funciones leerdatos(), mostrardatos()
# Crear 2 objetos ciudad1, ciudad2

class Ciudad:
  def __init__(self, nombre, pais, poblacion):
    self.nombre = nombre
    self.pais = pais
    self.poblacion = poblacion

  def leerdatos(self):
      self.nombre = input("Ingrese el nombre de la ciudad: ")
      self.pais = input("Ingrese el pais de la ciudad: ")
      self.poblacion = input("Ingrese la poblacion de la ciudad: ")

  def mostrardatos(self):
      print("Nombre de la ciudad: ", self.nombre)
      print("Pais de la ciudad: ", self.pais)
      print("Poblacion de la ciudad: ", self.poblacion)
      return self.nombre, self.pais, self.poblacion

# fin de la clase Ciudad

ciudad1 = Ciudad()
ciudad1.leerdatos()
ciudad1.mostrardatos()
print("*-*-*-*-*-*-*-*")

ciudad2 = Ciudad()
ciudad2.leerdatos()
ciudad2.mostrardatos()
print("*-*-*-*-*-*-*-*")

print("Software elaborado por Cepeda Darrell NL = 10")
```
